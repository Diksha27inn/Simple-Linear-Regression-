### **Step 1: Importing Python library**

```python
# Importing all libraries required in this notebook
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline
```

### **Step 2: Importing the Dataset**

```python
# Reading data from remote link
url = "http://bit.ly/w-data"
s_data = pd.read_csv(url)
print("Data imported successfully")

s_data.head(10)
```

### **Step 3: Plotting the Graph**

Let's plot our data points on 2-D graph to eyeball our dataset and see if we can manually find any relationship between the data.

**By using plot()function of the pandas dataframe, we will pass it to the column names for x-coordinate and y-coordinate. We can create the plot with the following script:**

```python
# Plotting the distribution of scores
s_data.plot(x='Hours', y='Scores', style= 'o')  
plt.title('Hours vs Percentage')  
plt.xlabel('Hours Studied')  
plt.ylabel('Percentage Score')  
plt.show()

# Making a Pie Graph
s_data.plot.pie(x = 'Hours', y = 'Scores')
plt.title('Pie Graph Representation')

# Making Bar Graph
s_data.plot.bar(x = 'Hours', y = 'Scores')
```

**From the graph above, we can clearly see that there is a positive linear relation between the number of hours studied and percentage of score.**

### Step 4: **Preparing the data**

The next step is to divide the data into "attributes" (inputs) and "labels" (outputs).

```python
X = s_data.iloc[:, :-1].values  
y = s_data.iloc[:, 1].values  
```

**Now that we have our attributes and labels, the next step is to split this data into training and test sets. We'll do this by using Scikit-Learn's built-in train_test_split() method:**

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                            test_size=0.2, random_state=0) 
```

### Step 5: **Training the Algorithm**
We have split our data into training and testing sets, and now is finally the time to train our algorithm by using **Linear Regression**.

Linear Regression = Predict Value of one variable by using another

```python
from sklearn.linear_model import LinearRegression  
regressor = LinearRegression()


regressor.fit(X_train, y_train) 

print("Training complete.")
```

### Step 6: **Plotting the regression line**

```python
# Plotting the regression line
line = regressor.coef_*X+regressor.intercept_
plt.title('Linear Regression vs Trained Model')

# Plotting for the test data
plt.scatter(X, y)
plt.plot(X, line);
plt.show()
```

### Step 7: **Making Predictions**
Now that we have trained our algorithm, it's time to make some predictions.

```python
print(X_test) # Testing data - In Hours

print('Score Prediction')
y_pred = regressor.predict(X_test) # Predicting the scores
print(y_pred)
```

### Step 8: Comparing Datasets

```python
# Comparing Actual vs Predicted
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df 

df.plot(kind= 'bar')
plt.title('Actual vs Predicted Values')
```

### Step 9: Making Predictions

```python
# You can also test with your own data
hours = [[9.25]]
own_pred = regressor.predict(hours)
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0]))
```

### **Evaluating the model**

The final step is to evaluate the performance of algorithm. This step is particularly important to compare how well different algorithms perform on a particular dataset. For simplicity here, we have chosen the mean square error. There are many such metrics.

```python
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_pred)) 
```
